```python
import pandas as pd
import sqlite3 as sql
```

```python
database = "jobs.db"
con = sql.connect(database)
```

```python
pd.read_sql("select * from sqlite_master",con)
```

```python
q = "select * from recent_grads"
pd.read_sql_query(q,con)
```

```python
q = """select sample_size from recent_grads where (Major = "PETROLEUM ENGINEERING" and women = 282)"""

pd.read_sql_query(q,con)
```

```python
q = """select sample_size from recent_grads where not (Major = "PETROLEUM ENGINEERING" and not sample_size < 5)"""

pd.read_sql_query(q,con)
```

```python
q = "select distinct(Major_category) from recent_grads"
pd.read_sql_query(q,con)
```

```python
q = "select distinct(Major_category) from recent_grads order by unemployed"
pd.read_sql_query(q,con)
```

```python
q = "select distinct(Major_category) from recent_grads order by unemployed desc"
pd.read_sql_query(q,con)
```

```python
q = "select major from recent_grads where major is null"
pd.read_sql_query(q,con)
```
